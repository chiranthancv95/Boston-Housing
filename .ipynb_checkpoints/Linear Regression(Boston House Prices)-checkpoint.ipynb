```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston



%matplotlib inline
```

```python
features=pd.DataFrame(boston.data, columns=boston.feature_names)
target=pd.DataFrame(boston.target, columns=['TARGET'])
```

```python
data=pd.concat([features, target],axis=1)
```

```python
data
```

```python
print(boston.DESCR)
```

```python
sep=0.8 * len(data)
sep=int(sep)
x_train=np.array(data['RM'][:sep])
y_train=np.array(data['TARGET'][:sep])

x_test=np.array(data['RM'][sep:])
y_test=np.array(data['TARGET'][sep:])
```

```python
plt.plot(x_train,y_train,'r.')
plt.xlabel("No of Rooms")
plt.ylabel("Price")
```

```python
def hypothesis(a,x,b):
    return a * x + b
```

```python
def error(a,x,b,y):
    e=0
    m=len(y)
    for i in range(m):
        e += np.power((hypothesis(a,x[i],b)-y[i]),2)
        
    return (1/(2*m)) *e
          
```

```python
def step_gradient(a,x,b,y, learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    for i in range(m):
        grad_a += 2/m * hypothesis(a,x[i],b - y[i]) *x[i]
        grad_b += 2/m * hypothesis(a,x[i],b - y[i])
        
    a=a - (grad_a * learning_rate)
    b=b - (grad_b * learning_rate)
    return a,b
```

```python
def descend(initial_a, initial_b,x,y, learning_rate, iterations):
    a=initial_a
    b=initial_b
    
    for i in range(iterations):
        a,b=step_gradient(a,x,b,y,learning_rate)
        e=error(a,x,b,y)
        if i %1000==0:
            print(f"Error: {e}, a:{a},b:{b}")
    return a,b
    
```

```python
init_a=0
init_b=0           

iters=10000
learning_rate=0.01
```

```python
final_a, final_b =descend(init_a, init_b, x_train, y_train, learning_rate, iters)
```

```python
plt.plot(x_train, y_train, 'r.', x_train,hypothesis(init_a, x_train, init_b),'b')
plt.plot(x_train, y_train, 'r.', x_train,hypothesis(final_a, x_train, final_b),'g')
```
